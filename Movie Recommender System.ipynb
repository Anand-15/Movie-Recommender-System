```python
#Importing libraries
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
#Now we will filter out the required columns for building Recommender system
#genres
#movie_id
#keywords
#overview
#title
#cast
#crew
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head(1)
```

```python
import ast
```

```python
movies["genres"][0] #Since it is a string of list we need to use ast to evaluate and convert it into list
```

```python
def convert(text):
    L=[]
#     print(ast.literal_eval(text))
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
convert(movies["genres"][0])
```

```python
#Lets check if there are any null values
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies["genres"]=movies["genres"].apply(convert)
movies.head()
```

```python
movies["keywords"][0] #Similar operation would be done on column keywords
```

```python
movies["keywords"]=movies["keywords"].apply(convert)
movies.head()
```

```python
movies["keywords"][0]
```

```python
movies['cast'][0]
```

```python
def convert_top_4_cast(text):
    L=[]
    count=0
#     print(ast.literal_eval(text))
    for i in ast.literal_eval(text):
        if count!=4:
            L.append(i['name'])
            count+=1
        else:
            break
    return L
```

```python
convert_top_4_cast(movies['cast'][0])
```

```python
movies["cast"]=movies["cast"].apply(convert_top_4_cast)
movies.head()
```

```python
movies["crew"][0]
```

```python
def fetch_director(text):
    L=[]
#     print(ast.literal_eval(text))
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
fetch_director(movies["crew"][0])
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies.sample(10)
```

```python
def collapse(L):
    L1=[]
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies["genres"]=movies["genres"].apply(collapse)
movies["keywords"]=movies["keywords"].apply(collapse)
movies["cast"]=movies["cast"].apply(collapse)
movies["crew"]=movies["crew"].apply(collapse)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x : x.split())
```

```python
movies.head()
```

```python
movies["tags"]=movies["overview"]+movies["genres"]+movies["keywords"]+movies["cast"]+movies["crew"]
```

```python
movies.head()
```

```python
new=movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
vector[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python

similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
movies.sample(4)
```

```python
recommend('Troy')
```
